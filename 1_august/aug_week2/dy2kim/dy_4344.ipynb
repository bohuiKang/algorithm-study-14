```python
c = int(input()) #테스트 케이스 개수
for k in range(c):
  lst = list(map(int, input().split()))

  sum = 0
  for i in range(1, lst[0]+1): #lst[0]은 명수이기에 1부터 시작
    sum += lst[i]
  mean = sum / lst[0] #전체 합산 / 명수 = 평균

  num = 0
  #평균을 넘긴 명수
  for j in range(1, lst[0]+1):
    if mean < lst[j]:
      num += 1

  print(f'{num/lst[0]*100:.3f}%')
```
