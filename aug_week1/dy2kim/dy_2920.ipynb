```python
ascen= list(map(int, '1 2 3 4 5 6 7 8'.split()))
answer= list(map(int,input().split()))

ascending = 0
descending = 0
for i in range(8):
    if answer[i] == ascen[i]:
      ascending += 1
    elif answer[-(i+1)] == ascen[i]:
      descending += 1

if ascending == 8:
  print("ascending")
elif descending == 8:
  print("descending")
else:
  print("mixed")
```
